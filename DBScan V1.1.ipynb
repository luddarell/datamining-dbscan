```python
def _euclidean(x,y):
    total=np.sum((x-y)**2) ** 0.5
    return total

def _minskowski(x,y):
    total=np.sum(abs(x-y))
    return total  
```

```python
def getNeighbors(X,P,eps):
    
    #X : Semua point yg ada di dataset
    #P : Point yg ditunjuk sebagai patokan (pusat untuk menghitung distance ke point lain)
    #eps : radius 
    seeds=[]
    for i in range(0,len(X)):
#         for j in range(len(X[i])):
        distance = _minskowski(P,X[i])
        if distance<eps and distance!=0:
            seeds.append(i)
    return np.array(seeds)

#_region_query(np.array([[65, 70], [80, 90]]), np.array([65, 70]), 0.3)

def expandCluster(X,NeighborPts,C,eps,minPts):
    
    #X: Semua point yg ada pada dataset
    #NeighborsPts : semua index yg merupakan ketetanggaan hasil dari perhitungan minskowski
    #C : Label untuk cluster baru 
    i=0
    while i <len(NeighborPts):
        next_point=NeighborPts[i]
        if X[next_point,-1]==-1:
            X[next_point,-1]=C
            
        elif X[next_point,-1]==0:
            X[next_point,-1]=C
            
            next_point_neighborPts=getNeighbors(X[:,:-1],X[next_point,:-1],eps)
            
            if len(next_point_neighborPts)>=minPts:
                print("Next Neighbor", next_point_neighborPts)
                print(X)
                NeighborPts= np.unique(np.concatenate((NeighborPts,next_point_neighborPts)))
                print(NeighborPts)
        i+=1
```

```python
def dbscan(X,eps,minPts):
#     labels=[0]*len(X)
    C=0
    
    listClust = X[:,-1] # Get reference to all cluster point
    while np.count_nonzero(listClust) < len(listClust): # Selama count non zero < panjang list cluster, berarti masih ada nilai 0
        randInt = np.random.randint(0, len(listClust)) # generate random index
        print("start index", randInt)
        if not (X[randInt,-1]==0):
            continue
        
        NeighborPts=getNeighbors(X[:,:-1],X[randInt,:-1],eps)
        print("First Neighbor", NeighborPts)
        
        if len(NeighborPts)<minPts:
            X[randInt,-1]=-1
        else :
            C+=1
            X[randInt,-1] = C
            print(X)
            expandCluster(X,NeighborPts,C,eps,minPts)
            
        print('Cluster Result\n', X, '\n')
    return X
        
        
```
